```python
import cv2
import numpy as np
import picamera
import picamera.array
import time

# Define the detectobj function
def detectobj(image):
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    lower_green = np.array([45, 50, 50])
    upper_green = np.array([75, 255, 255])
    lower_blue = np.array([100, 50, 50])
    upper_blue = np.array([130, 255, 255])
    mask_green = cv2.inRange(hsv, lower_green, upper_green)
    mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)

    green_detected = True
    blue_detected = True

    # Calculate the percentage of green color in the image
    total_pixels = np.size(mask_green)
    green_pixels = cv2.countNonZero(mask_green)
    green_percentage = (green_pixels / total_pixels) * 100

    # Print the result
    print('Green color percentage:', green_percentage, '%')

    # Calculate the percentage of blue color in the image
    total_pixels = np.size(mask_blue)
    blue_pixels = cv2.countNonZero(mask_blue)
    blue_percentage = (blue_pixels / total_pixels) * 100

    # Print the result
    print('Blue color percentage:', blue_percentage, '%','\n')

    # Check if green color is detected
    if green_percentage < 1.0:
        green_detected = False

    # Check if blue color is detected
    if blue_percentage < 1.0:
        blue_detected = False

    # Print the result
    print('Green color detected:', green_detected)

    # Print the result
    print('Blue color detected:', blue_detected)
    
    full_slots = 0

    if green_detected == True and blue_detected==True :
        full_slots=2
    elif green_detected== True or blue_detected==True :
        full_slots=1


    # Combine the green and purple masks
    mask = cv2.bitwise_or(mask_green, mask_blue)

    # Apply the mask to the original image
    result = cv2.bitwise_and(image, image, mask=mask)
    
    return full_slots, result


# Initialize the camera
camera = picamera.PiCamera()

# Set the resolution of the camera
camera.resolution = (640, 480)
#camera.framerate=30
time.sleep(2)


# Start the preview to initialize the camera
camera.start_preview()

# Create an array to store the captured frames
frame_array = picamera.array.PiRGBArray(camera)

# Capture video from the camera and process the frames
for frame in camera.capture_continuous(frame_array, format='bgr', use_video_port=True):
    # Get the image data from the captured frame
    image = frame.array

    # Process the image using the detectobj function
    full_slots, result = detectobj(image)

    # Display the processed image on the screen
    cv2.imshow('Video', result)

    # Wait for a key press and check if the 'q' key was pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    # Clear the frame array for the next frame
    frame_array.truncate(0)
    #time.sleep(5)
# Release the camera and close the window
camera.close()
cv2.destroyAllWindows()
```
